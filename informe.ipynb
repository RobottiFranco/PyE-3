# Informe sobre Distribuciones Discretas

## Introducción

En este informe, exploraremos tres distribuciones discretas comunes: la distribución binomial, la distribución geométrica y la distribución de Poisson. Analizaremos sus definiciones matemáticas, sus características estadísticas y sus aplicaciones en contextos prácticos.

## Distribución Binomial

### Definición Matemática:
La distribución binomial modela el número de éxitos en una secuencia de \( n \) ensayos independientes, donde cada ensayo tiene dos posibles resultados: éxito con probabilidad \( p \) y fracaso con probabilidad \( 1-p \). La función de masa de probabilidad (PMF) de la distribución binomial está dada por:

\[ P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} \]

donde \( k \) es el número de éxitos, \( n \) es el número total de ensayos y \( p \) es la probabilidad de éxito en cada ensayo.

### Características Estadísticas:
- **Media**: \( \mu = np \)
- **Varianza**: \( \sigma^2 = np(1-p) \)
- **Moda**: Depende de los parámetros \( n \) y \( p \), pero generalmente se encuentra cerca de \( np \).

### Usos Comunes:
- Modelado de eventos binarios como éxito/fracaso, sí/no.
- Experimentos de conteo con un número fijo de ensayos.

## Distribución Geométrica

### Definición Matemática:
La distribución geométrica modela el número de ensayos independientes de Bernoulli necesarios antes de que ocurra el primer éxito. La función de masa de probabilidad (PMF) de la distribución geométrica está dada por:

\[ P(X = k) = (1-p)^{k-1} p \]

donde \( k \) es el número de ensayos antes del primer éxito y \( p \) es la probabilidad de éxito en cada ensayo.

### Características Estadísticas:
- **Media**: \( \mu = \frac{1}{p} \)
- **Varianza**: \( \sigma^2 = \frac{1-p}{p^2} \)
- **Moda**: \( 1 \), ya que el primer ensayo siempre se cuenta.

### Usos Comunes:
- Modelado de la cantidad de tiempo o ensayos necesarios para que ocurra un evento específico.
- Análisis de fallos o tiempos de espera.

## Distribución de Poisson

### Definición Matemática:
La distribución de Poisson modela el número de eventos que ocurren en un intervalo de tiempo o espacio fijo, con una tasa media de ocurrencia \( \lambda \). La función de masa de probabilidad (PMF) de la distribución de Poisson está dada por:

\[ P(X = k) = \frac{{e^{-\lambda} \lambda^k}}{{k!}} \]

donde \( k \) es el número de eventos y \( \lambda \) es la tasa media de ocurrencia.

### Características Estadísticas:
- **Media**: \( \mu = \lambda \)
- **Varianza**: \( \sigma^2 = \lambda \)
- **Moda**: \( \lfloor \lambda \rfloor \), el entero más cercano a \( \lambda \).

### Usos Comunes:
- Modelado de la cantidad de llegadas de clientes, llamadas telefónicas, errores de impresión, etc., en un intervalo de tiempo o espacio fijo.
- Procesos de Poisson en teoría de colas y procesos estocásticos.
