## Distribucin Binomial

### Definicin Matemtica:
La distribucin binomial modela el nmero de xitos en una secuencia de \( n \) ensayos independientes, donde cada ensayo tiene dos posibles resultados: xito con probabilidad \( p \) y fracaso con probabilidad \( 1-p \). La funcin de masa de probabilidad (PMF) de la distribucin binomial est dada por:

\[ P(X = k) = \binom{n}{k} p^k (1-p)^{n-k} \]

donde \( k \) es el nmero de xitos, \( n \) es el nmero total de ensayos y \( p \) es la probabilidad de xito en cada ensayo.

### Caractersticas Estadsticas:
- **Media**: \( \mu = np \)
- **Varianza**: \( \sigma^2 = np(1-p) \)
- **Moda**: Depende de los parmetros \( n \) y \( p \), pero generalmente se encuentra cerca de \( np \).

### Usos Comunes:
- Modelado de eventos binarios como xito/fracaso, s/no.
- Experimentos de conteo con un nmero fijo de ensayos.

## Distribucin Geomtrica

### Definicin Matemtica:
La distribucin geomtrica modela el nmero de ensayos independientes de Bernoulli necesarios antes de que ocurra el primer xito. La funcin de masa de probabilidad (PMF) de la distribucin geomtrica est dada por:

\[ P(X = k) = (1-p)^{k-1} p \]

donde \( k \) es el nmero de ensayos antes del primer xito y \( p \) es la probabilidad de xito en cada ensayo.

### Caractersticas Estadsticas:
- **Media**: \( \mu = \frac{1}{p} \)
- **Varianza**: \( \sigma^2 = \frac{1-p}{p^2} \)
- **Moda**: \( 1 \), ya que el primer ensayo siempre se cuenta.

### Usos Comunes:
- Modelado de la cantidad de tiempo o ensayos necesarios para que ocurra un evento especfico.
- Anlisis de fallos o tiempos de espera.

## Distribucin de Poisson

### Definicin Matemtica:
La distribucin de Poisson modela el nmero de eventos que ocurren en un intervalo de tiempo o espacio fijo, con una tasa media de ocurrencia \( \lambda \). La funcin de masa de probabilidad (PMF) de la distribucin de Poisson est dada por:

\[ P(X = k) = \frac{{e^{-\lambda} \lambda^k}}{{k!}} \]

donde \( k \) es el nmero de eventos y \( \lambda \) es la tasa media de ocurrencia.

### Caractersticas Estadsticas:
- **Media**: \( \mu = \lambda \)
- **Varianza**: \( \sigma^2 = \lambda \)
- **Moda**: \( \lfloor \lambda \rfloor \), el entero ms cercano a \( \lambda \).

### Usos Comunes:
- Modelado de la cantidad de llegadas de clientes, llamadas telefnicas, errores de impresin, etc., en un intervalo de tiempo o espacio fijo.
- Procesos de Poisson en teora de colas y procesos estocsticos.
